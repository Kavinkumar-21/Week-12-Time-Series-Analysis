```python
def get_pdf_probability(dataset, start_range, end_range):
    
    from scipy.stats import norm
    
    sns.histplot(dataset, kde=True, color="green")

    plt.axvline(start_range,color="red")
    plt.axvline(end_range,color="red")

    mean = dataset.mean()
    std = dataset.std()

    print("Mean =",mean)
    print("Std =",std)

    prob = norm.cdf(end_range,mean,std) - norm.cdf(start_range,mean,std)

    return prob
```

```python
get_pdf_probability(stk_data["Low"],219,230)
```

```python
def stdNBgraph(dataset):

    import seaborn as sns
    import matplotlib.pyplot as plt

    dataset = dataset.dropna()

    z_score = (dataset - dataset.mean()) / dataset.std()

    sns.histplot(z_score, kde=True)

    plt.title("Standard Normal Distribution")
    plt.show()

    return z_score.mean()
```

```python
stdNBgraph(stk_data["Low"])
```

```python
import pandas as pd

def freqTable(columnName, dataset):

    value_counts = dataset[columnName].value_counts()

    frq = pd.DataFrame({
        "Unique_Values": value_counts.index,
        "Frequency": value_counts.values
    })

    frq["Relative_Fre"] = (frq["Frequency"] / len(dataset[columnName])) * 100
    frq["Cumulative"] = frq["Relative_Fre"].cumsum()

    return frq
```

```python
freqTable("Low", stk_data)
```
